```python
%matplotlib inline

import pandas as pd
import matplotlib.pyplot as plt
import datetime

plt.rc('figure', figsize=(12, 8))
```

```python
april1 = pd.read_table("ITS_Primero/N. Aquatic Park_downstream_PM_5.25/occ/April/pems.dot.ca.gov11.txt")
print(april1.head())
april1.columns=['date', 'L1_flow', 'L1_occ', 'L2_flow', 'L2_occ','L3_flow', 'L3_occ','L4_flow', 'L4_occ', 'L5_flow', 'L5_occ', 'flow', 'occ', 'lanes', 'observed']
april1.date = pd.to_datetime(april1.date)
```
