# Análisis de accidentes

```python
%matplotlib inline

import pandas as pd
import matplotlib.pyplot as plt
import datetime
import os
import matplotlib.dates as mdates
import numpy as np
from collections import Counter

CRASH_PATH = '/home/josejm/transportes/ITS_Primero/HSIS 2006 crash data.xlsx'
```

```python
crash = pd.read_excel(CRASH_PATH)
crash.acc_date = pd.to_datetime(crash.acc_date, format="%Y%m%d")
crash.caseno = crash.caseno.map(lambda item: pd.to_datetime(item[:12],format="%Y%m%d%H%M", errors='coerce' ))
crash.head()
```

```python
crash.columns
```

## Número de vehículos involucrados en accidentes

```python
plt.hist(crash.numvehs)
Counter(crash.numvehs)
```

## Número de vehículos involucrados por fecha DIA/MES/AÑO

```python
plt.plot(crash.acc_date, crash.numvehs, 'ro')
```

#### Filtro las medidas anteriores a Junio/2006

```python
crash = crash[crash.acc_date > datetime.datetime(2006, 6, 1)]
```

```python
plt.plot(crash.acc_date, crash.numvehs, 'ro')
```

## Accidentes por hora del día

```python
crash = crash[crash.caseno.notnull()]
times = crash.caseno.map(lambda t: t.replace(day=1, month=1, year=2000))

times.hist(figsize=(15,5), bins=288)
```

```python
hours = crash.caseno.map(lambda t: t.hour)
c = Counter(hours)
hours_df = pd.DataFrame.from_dict(c, orient='index')
hours_ser = pd.Series(hours_df.sort_index().T.values[0])
plt.bar(np.arange(24), hours_ser)
```

## Heridos/Muertos en función del día del año/día de la semana

```python
plt.rc('figure', figsize=(16, 8))
plt.plot(crash.acc_date, crash.tot_inj, 'ro')
```

```python
fig, ax = plt.subplots(1, 1)

date_inj = crash[['acc_date', 'tot_inj']]
pordia_inj = date_inj.groupby('acc_date').sum()
pordia_inj.plot(ax=ax)

date_kill = crash[['acc_date', 'tot_kill']]
pordia_kill = date_kill.groupby('acc_date').sum()
pordia_kill.plot(ax=ax)

out = crash['acc_date'].to_frame()
out['accidentes'] = 1
pordia_acc = out.groupby('acc_date').sum()
pordia_acc.plot(ax=ax)
```

```python
fig, ax = plt.subplots(1, 1)

out = crash['weekday'].to_frame()
out['accidentes'] = 1
pordia_acc = out.groupby('weekday').sum()
pordia_acc.index = ['D', 'L', 'M', 'X', 'J', 'V', 'S']
pordia_acc.plot.bar(ax=ax, color='Red')


date_inj = crash[['weekday', 'tot_inj']]
pordia_inj = date_inj.groupby('weekday').sum()
pordia_inj.index = ['D', 'L', 'M', 'X', 'J', 'V', 'S']
pordia_inj.plot(ax=ax)

date_kill = crash[['weekday', 'tot_kill']]
pordia_kill = date_kill.groupby('weekday').sum()
pordia_kill.index = ['D', 'L', 'M', 'X', 'J', 'V', 'S']
pordia_kill.plot(ax=ax)
```
